```python
import pandas as pd
import matplotlib.pyplot as plt

results = {
    'Model': ['LSTM', 'BERT (DistilBERT)'],
    'Test Accuracy': [81.87, 92.10],
    'Training Time': ['~17 min', '~90 min'],
    'Parameters': ['~5M', '~66M']
}

df_results = pd.DataFrame(results)
print(df_results)
```

```python
plt.bar(df_results['Model'], df_results['Test Accuracy'])
plt.xlabel('Model')
plt.ylabel('Test Accuracy')
plt.title('Comparison of Accuracy Models (LSTM vs BERT')
for i, (model, acc) in enumerate(zip(df_results['Model'], df_results['Test Accuracy'])):
    plt.text(i, acc, f'{acc}%', ha='center')
plt.show()
```
